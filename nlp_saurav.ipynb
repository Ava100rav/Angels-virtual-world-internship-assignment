```python
import pandas as pd
```

```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_excel('/content/drive/MyDrive/angels-virtual-world-internship/Headlines_5000.xlsx')
```

```python
df.head()
```

```python
df['scores'] = df['headline'].apply(lambda headline:sid.polarity_scores(headline))
```

```python
df.head()
```

```python
df.drop(['date', 'time_12hr', 'time_24hr', 'Unnamed: 5', 'Unnamed: 6'], axis = 1, inplace = True)
```

```python
df['scores'][0]
```

```python
df['compound'] = df['scores'].apply(lambda d:d['compound'])
```

```python
df['com_score'] = df['compound'].apply(lambda score: 'pos' if score >= 0 else 'neg')
```

```python
df.head()
```

```python
df.rename(columns={'com_score': 'sentiment', 'compound': 'score'}, inplace=True)
```

```python
df.drop(['category','scores'],axis=1,inplace=True)
```
